```python
from std_msgs.msg import Header
from trajectory_msgs.msg import JointTrajectory
from trajectory_msgs.msg import JointTrajectoryPoint
import rospy

print("Packages Loaded.")
```

```python
waypoints = [[-7.718218911456631e-05, -1.4999110556748372, 1.499937432019225, -1.499941062374841, -1.5000240680440013, -5.9393964622955764e-05], [0,-1.6,1.3,-1.5,-1.5,0]]
#waypoints_gripper =[[0.3998742830500399, 0.3998742830500399],[0.0,0.0]]


def main():

    
    rospy.init_node('send_joints')

    #arm_controller, JointTrajectory
    pub = rospy.Publisher('/arm_controller/command',
                          JointTrajectory,
                          queue_size=10)

    #gripper_controller, JointTrajectory	
    #gripper_pub=rospy.Publisher('/gripper_controller/command',
    #                      JointTrajectory,
    #                      queue_size=10)


    # Create the topic message
    traj = JointTrajectory()
    #traj2= JointTrajectory()

    traj.header = Header()
   # traj2.header = Header()

    # Joint names for UR5
    traj.joint_names = ['shoulder_pan_joint', 'shoulder_lift_joint','elbow_joint', 
                        'wrist_1_joint', 'wrist_2_joint',
                        'wrist_3_joint']

  #  traj2.joint_names=['gripper_finger1_joint', 'gripper_finger2_joint']

    rate = rospy.Rate(1)
    cnt = 0

    pts = JointTrajectoryPoint()
    #pts2= JointTrajectoryPoint()

    traj.header.stamp = rospy.Time.now()
    #traj2.header.stamp= rospy.Time.now()

    while not rospy.is_shutdown():

        cnt += 1
        if cnt%2==0:
                pts.positions = waypoints[0]

        else:
                pts.positions = waypoints[1]
        #   pts2.positions = waypoints_gripper[0]

        pts.time_from_start = rospy.Duration(1.0)
        #  pts2.time_from_start = rospy.Duration(1.0)
        # Set the points to the trajectory
        traj.points = []
        # traj2.points = []

        traj.points.append(pts)
        #traj2.points.append(pts)

        # Publish the message
        pub.publish(traj)
    # gripper_pub(traj2)
        rate.sleep()

if __name__ == '__main__':
    try:
        main()
    except rospy.ROSInterruptException:
        print ("Program interrupted before completion")
```
